# RAGAS Before-and-After Comparison

This notebook compares the baseline Networking RAG system (Phase 2) and the improved Networking RAG system (Phase 4) using the same RAGAS evaluation metrics.

```python
 import pandas as pd

 print(
        "Libraries imported successfully!"
 )
```

```python
# Load evaluation result files

try:

    phase2_df = pd.read_csv(
        "phase2_ragas_results.csv"
    )

    phase4_df = pd.read_csv(
        "phase4_ragas_results.csv"
    )

    print(
        "Files loaded successfully!"
    )

except Exception as e:

    print(
        "Loading Error:"
    )

    print(e)
```

```python
# Verify loaded files

try:

    print(
        "\nPhase 2 Results:"
    )

    print(
        phase2_df.head()
    )

    print(
        "\nPhase 4 Results:"
    )

    print(
        phase4_df.head()
    )

except Exception as e:

    print(e)
```

```python
# Calculate average metric scores

try:

    metrics = [

        "faithfulness",

        "answer_relevancy",

        "context_precision",

        "context_recall"

    ]

    phase2_scores = {}

    phase4_scores = {}

    for metric in metrics:

        phase2_scores[
            metric
        ] = phase2_df[
            metric
        ].mean()

        phase4_scores[
            metric
        ] = phase4_df[
            metric
        ].mean()

    print(
        "Average scores calculated successfully!"
    )

except Exception as e:

    print(e)
```

```python
# Display average scores

try:

    print(
        "\nPhase 2 Scores\n"
    )

    print(
        phase2_scores
    )

    print(
        "\nPhase 4 Scores\n"
    )

    print(
        phase4_scores
    )

except Exception as e:

    print(e)
```

```python
# Create before-and-after comparison table

try:

    comparison_df = pd.DataFrame({

        "Metric": [

            "Faithfulness",

            "Answer Relevancy",

            "Context Precision",

            "Context Recall"

        ],

        "Phase 2": [

            phase2_scores[
                "faithfulness"
            ],

            phase2_scores[
                "answer_relevancy"
            ],

            phase2_scores[
                "context_precision"
            ],

            phase2_scores[
                "context_recall"
            ]

        ],

        "Phase 4": [

            phase4_scores[
                "faithfulness"
            ],

            phase4_scores[
                "answer_relevancy"
            ],

            phase4_scores[
                "context_precision"
            ],

            phase4_scores[
                "context_recall"
            ]

        ]

    })

    comparison_df[
        "Improvement"
    ] = (

        comparison_df[
            "Phase 4"
        ]

        -

        comparison_df[
            "Phase 2"
        ]

    )

    print(
        comparison_df
    )

except Exception as e:

    print(e)
```

**Evaluation Summary**

The baseline (**Phase 2**) and improved (**Phase 4**) RAG systems were evaluated using **RAGAS metrics** on the same set of evaluation questions. A comparison table was created to measure the impact of **Better Chunking** and **Hybrid Search**.

**RAGAS Metrics**

• **Faithfulness** – Checks whether the generated answer is supported by the retrieved context.

• **Answer Relevancy** – Measures how well the answer addresses the user's question.

• **Context Precision** – Measures the relevance of the retrieved chunks.

• **Context Recall** – Measures whether all relevant information was retrieved.


**Results**

• **Faithfulness:** 0.9167 → 1.0000

• **Answer Relevancy:** 0.7908 → 0.8040

• **Context Precision:** 1.0000 → 0.9258

• **Context Recall:** 1.0000 → 1.0000

The improved system achieved better answer grounding and maintained perfect context recall. The slight decrease in **Context Precision** is expected because **Hybrid Search** retrieves a broader set of relevant chunks by combining vector search and BM25 retrieval, improving coverage while introducing a small amount of additional context.
