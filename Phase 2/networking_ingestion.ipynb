```python
try:
    topics = [
        "Transmission Control Protocol",
        "User Datagram Protocol",
        "Hypertext Transfer Protocol",
        "Hypertext Transfer Protocol Secure",
        "Domain Name System",
        "Internet Protocol",
        "OSI model",
        "TCP/IP model",
        "Routing",
        "Network switch"
    ]

    print("Topics loaded successfully!")
    print("Number of topics:", len(topics))

except Exception as e:
    print("Error creating topic list:")
    print(e)
```

```python
try:

    saved_count = 0

    for topic in topics:

        try:
            page = wiki.page(topic)

            if page.exists():

                filename = topic.replace("/", "_").replace(" ", "_") + ".txt"

                with open(filename, "w", encoding="utf-8") as file:
                    file.write(page.text)

                saved_count += 1
                print(f"Saved: {filename}")

            else:
                print(f"Page not found: {topic}")

        except Exception as article_error:
            print(f"Error processing {topic}:")
            print(article_error)

    print(f"\nSuccessfully saved {saved_count} files.")

except Exception as e:
    print("Error during article download process:")
    print(e)
```

```python
try:
    import os

    documents = []

    for file in os.listdir():
        if file.endswith(".txt"):
            with open(file, "r", encoding="utf-8") as f:
                documents.append(f.read())

    print("Documents loaded successfully!")
    print("Number of documents:", len(documents))

except Exception as e:
    print("Error loading documents:")
    print(e)
```

```python
try:
    from langchain_text_splitters import RecursiveCharacterTextSplitter

    print("Text splitter imported successfully!")

except Exception as e:
    print("Import Error:")
    print(e)
```

```python
try:
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=2500,
        chunk_overlap=50
    )

    chunks = []

    for document in documents:
        document_chunks = text_splitter.split_text(document)
        chunks.extend(document_chunks)

    print("Chunking completed successfully!")
    print("Total chunks created:", len(chunks))

except Exception as e:
    print("Chunking Error:")
    print(e)
```

```python
try:
    print("First Chunk:\n")
    print(chunks[0])

    print("\nChunk Length:")
    print(len(chunks[0]))

except Exception as e:
    print("Verification Error:")
    print(e)
```

```python
try:
    from google import genai
    from google.colab import userdata

    client = genai.Client(
        api_key=userdata.get("GEMINI_API_KEY")
    )

    print("Gemini client initialized successfully!")

except Exception as e:
    print("Gemini Client Error:")
    print(e)
```

```python
try:
    sample_chunk = chunks[0]

    response = client.models.embed_content(
        model="models/gemini-embedding-001",
        contents=sample_chunk
    )

    print("Embedding generated successfully!")
    print("Embedding dimension:", len(response.embeddings[0].values))

except Exception as e:
    print("Embedding Error:")
    print(e)
```

```python
try:
    import time

    chunk_embeddings = []

    for i, chunk in enumerate(chunks):

        try:
            response = client.models.embed_content(
                model="models/gemini-embedding-001",
                contents=chunk
            )

            chunk_embeddings.append(
                response.embeddings[0].values
            )

            if (i + 1) % 10 == 0:
                print(f"Processed {i + 1} chunks")

            # Delay to avoid rate limits
            time.sleep(1)

        except Exception as chunk_error:
            print(f"Error processing chunk {i + 1}:")
            print(chunk_error)

            print("Waiting 35 seconds before retrying...")
            time.sleep(35)

    print("\nEmbedding generation completed!")
    print("Total embeddings generated:", len(chunk_embeddings))

except Exception as e:
    print("Embedding Generation Error:")
    print(e)
```

```python
try:
    import pickle

    with open("chunk_embeddings.pkl", "wb") as f:
        pickle.dump(chunk_embeddings, f)

    with open("chunks.pkl", "wb") as f:
        pickle.dump(chunks, f)

    print("Embeddings and chunks saved successfully!")

except Exception as e:
    print("Error saving files:")
    print(e)
```
