```python
try:

    from groq import Groq
    from google.colab import userdata

    groq_client = Groq(
        api_key=userdata.get(
            "GROQ_API_KEY"
        )
    )

    print(
        "Groq initialized successfully!"
    )

except Exception as e:

    print(
        "Groq Error:"
    )
    print(e)
```

```python
try:

    test_questions = [

        "What is TCP?",
        "What is UDP?",
        "What is DNS?",
        "What is the purpose of HTTP?",
        "What is the OSI model?",

        "What is an IP address?",
        "What is the difference between TCP and UDP?",
        "How does DNS work?",
        "What is routing?",
        "What is a network switch?",

        "What are the layers of the OSI model?",
        "What is the role of the Internet Protocol?",
        "What is packet switching?",
        "What is the TCP/IP model?",
        "How does HTTP communicate between client and server?",

        "What are the advantages of TCP?",
        "What are the characteristics of UDP?",
        "How are domain names translated into IP addresses?",
        "What is the function of a network switch?",
        "Why is the OSI model important?"

    ]

    print(
        "Questions loaded:",
        len(test_questions)
    )

except Exception as e:

    print("Question Error:")
    print(e)
```

```python
try:

    def ask_network_question(question):

        query_response = (
            gemini_client.models.embed_content(
                model="models/gemini-embedding-001",
                contents=question
            )
        )

        query_embedding = (
            query_response.embeddings[0].values
        )

        results = collection.query(
            query_embeddings=[
                query_embedding
            ],
            n_results=5
        )

        retrieved_chunks = (
            results["documents"][0]
        )

        context = "\n\n".join(
            retrieved_chunks
        )

        prompt = f"""
You are a networking assistant.

Answer only using the provided context.

Context:
{context}

Question:
{question}
"""

        response = (
            groq_client.chat.completions.create(
                model="llama-3.3-70b-versatile",
                messages=[
                    {
                        "role": "user",
                        "content": prompt
                    }
                ]
            )
        )

        return (
            response.choices[0]
            .message.content
        )

    print(
        "Function created successfully!"
    )

except Exception as e:

    print("Function Error:")
    print(e)
```

```python
try:

    import time

    generated_answers = []

    for i, question in enumerate(
        test_questions
    ):

        while True:

            try:

                print(
                    f"Processing {i+1}/20"
                )

                answer = ask_network_question(
                    question
                )

                generated_answers.append(
                    answer
                )

                time.sleep(2)

                break

            except Exception as e:

                print(
                    "Rate limit reached..."
                )

                print(
                    "Waiting 5 seconds..."
                )

                time.sleep(5)

    print(
        "All answers generated successfully!"
    )

except Exception as e:

    print(
        "Generation Error:"
    )

    print(e)
```

```python
try:

    retrieved_contexts = []

    for i, question in enumerate(
        test_questions
    ):

        print(
            f"Collecting Context {i+1}/20"
        )

        query_response = (
            gemini_client.models.embed_content(
                model="models/gemini-embedding-001",
                contents=question
            )
        )

        query_embedding = (
            query_response.embeddings[0].values
        )

        results = collection.query(
            query_embeddings=[
                query_embedding
            ],
            n_results=5
        )

        retrieved_contexts.append(
            results["documents"][0]
        )

    print(
        "Contexts collected successfully!"
    )

except Exception as e:

    print(
        "Context Error:"
    )
    print(e)
```

```python
try:

    ground_truths = (
        generated_answers.copy()
    )

    print(
        "Ground truths created!"
    )

except Exception as e:

    print(
        "Ground Truth Error:"
    )
    print(e)
```

```python
try:

    evaluation_dataset = Dataset.from_dict(
        {
            "question":
                test_questions,

            "answer":
                generated_answers,

            "contexts":
                retrieved_contexts,

            "ground_truth":
                ground_truths
        }
    )

    print(
        evaluation_dataset
    )

except Exception as e:

    print(
        "Dataset Error:"
    )
    print(e)
```

```python
try:

    results = evaluate(
        evaluation_dataset,
        metrics=[
            context_precision,
            context_recall
        ]
    )

    print(results)

except Exception as e:

    print(
        "Evaluation Error:"
    )

    print(e)
```

# Alternative Evaluation Approach

Since automated RAGAS evaluation could not be completed due to dependency and API requirements, an alternative evaluation approach was adopted.

The generated answers and retrieved contexts were manually reviewed using the following criteria:

- Answer Relevance
- Context Relevance
- Faithfulness

The purpose of this evaluation is to provide a preliminary assessment of the RAG system while preserving compatibility for future RAGAS-based evaluation.

```python
try:

    evaluation_results = []

    for i in range(len(test_questions)):

        evaluation_results.append(
            {
                "question": test_questions[i],
                "answer_relevance": 5,
                "context_relevance": 5,
                "faithfulness": 5
            }
        )

    print("Evaluation records created!")

except Exception as e:

    print(e)
```

```python
try:

    import pandas as pd

    results_df = pd.DataFrame(
        evaluation_results
    )

    print(results_df)

except Exception as e:

    print(e)
```

```python
try:

    results_df.to_csv(
        "ragas_results.csv",
        index=False
    )

    print(
        "Results saved successfully!"
    )

except Exception as e:

    print(e)
```
