```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = [
    "TCP is a reliable transport protocol",
    "UDP provides fast communication",
    "HTTP is used for websites",
    "HTTPS is secure web communication",
    "The OSI model has seven layers",
    "Routing determines packet paths",
    "Switches operate at the data link layer",
    "IP addresses identify devices on a network",
    "DNS converts domain names into IP addresses",
    "Firewalls improve network security",

    "Machine learning uses training data",
    "Neural networks are inspired by the human brain",
    "Artificial intelligence is transforming industries",
    "Deep learning is a subset of machine learning",
    "Python is widely used in AI applications",

    "Cats drink milk",
    "Dogs bark loudly",
    "Birds can fly in the sky",
    "Apples are healthy fruits",
    "The sun rises in the east"
]
```

```python
try:
    embeddings = model.encode(sentences)

    print("Embeddings generated successfully!")
    print("Shape of embeddings:", embeddings.shape)

except Exception as e:
    print("Error while generating embeddings:")
    print(e)
```

```python
try:
    query = "Explain TCP protocol"

    query_embedding = model.encode([query])

    from sklearn.metrics.pairwise import cosine_similarity

    similarities = cosine_similarity(query_embedding, embeddings)

    top_5_indices = similarities[0].argsort()[-5:][::-1]

    print("Query:", query)
    print("\nTop 5 Similar Sentences:\n")

    for idx in top_5_indices:
        print(sentences[idx])
        print("Similarity Score:", similarities[0][idx])
        print()

except Exception as e:
    print("Error during similarity search:")
    print(e)
```
