```python
from google import genai
from google.colab import userdata
import numpy as np
```

```python
client = genai.Client(
    api_key=userdata.get("GEMINI_API_KEY")
)
```

```python
sentences = [
    "TCP is a reliable transport protocol",
    "UDP provides fast communication",
    "HTTP is used for websites",
    "HTTPS is secure web communication",
    "The OSI model has seven layers",
    "Routing determines packet paths",
    "Switches operate at the data link layer",
    "IP addresses identify devices on a network",
    "DNS converts domain names into IP addresses",
    "Firewalls improve network security",
    "Machine learning uses training data",
    "Neural networks are inspired by the human brain",
    "Artificial intelligence is transforming industries",
    "Deep learning is a subset of machine learning",
    "Python is widely used in AI applications",
    "Cats drink milk",
    "Dogs bark loudly",
    "Birds can fly in the sky",
    "Apples are healthy fruits",
    "The sun rises in the east"
]
```

```python
try:
    embeddings = []

    for sentence in sentences:
        response = client.models.embed_content(
            model="gemini-embedding-001",
            contents=sentence
        )

        embeddings.append(response.embeddings[0].values)

    print("Embeddings generated successfully!")
    print("Number of embeddings:", len(embeddings))
    print("Embedding dimension:", len(embeddings[0]))

except Exception as e:
    print("Error while generating embeddings:")
    print(e)
```

```python
embeddings = np.array(embeddings)
print("Embeddings shape:", embeddings.shape)
```

```python
try:
    query = "Explain TCP protocol"

    query_response = client.models.embed_content(
        model="gemini-embedding-001",
        contents=query
    )

    query_embedding = np.array(query_response.embeddings[0].values)

    print("Query embedding generated successfully!")

except Exception as e:
    print("Error while generating query embedding:")
    print(e)
```

```python
try:
    from sklearn.metrics.pairwise import cosine_similarity

    similarities = cosine_similarity(
        [query_embedding],
        embeddings
    )

    top_5_indices = similarities[0].argsort()[-5:][::-1]

    print("Query:")
    print(query)

    print("\nTop 5 Similar Sentences:\n")

    for idx in top_5_indices:
        print(sentences[idx])
        print("Similarity Score:", similarities[0][idx])
        print()

except Exception as e:
    print("Error during similarity search:")
    print(e)
```
